```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC




```

```python
#  Representations des données
data = np.array([[2,3], [4,2], [3,5], [6,6], [7,7], [5,6]])
labels = ['A', 'A', 'A', 'B', 'B', 'B']


markers = {'A': 'o', 'B': 'x'}
colors = {'A': 'blue', 'B': 'red'}

for (x, y), label in zip(data, labels):
    plt.scatter(x, y, marker=markers[label], color=colors[label])


plt.xlabel("X")
plt.ylabel("Y")
plt.title("Representation des données")
plt.show()
```

```python
# Entrainement du jeu de données
svm = SVC(kernel='linear')
svm.fit(data, labels)
```

```python
# Détermination de la droite de séparation

coef_a = svm.coef_
w1 = coef_a[0][0]
w2 = coef_a[0][1]
intercept = svm.intercept_


print(w1)
print(w2)
print(intercept)

# w1*x + w2*y + b = 0
def straight(x_):
    return (-x_*w1 - intercept)/w2

def straight_left(x_):
    return (-1-x_*w1 - intercept)/w2

def straight_right(x_):
    return (1-x_*w1 - intercept)/w2

for (x, y), label in zip(data, labels):
    plt.scatter(x, y, marker=markers[label], color=colors[label], label=label)

plt.plot([straight(x) for x in range(10)])
plt.plot([straight_left(x) for x in range(10)])
plt.plot([straight_right(x) for x in range(10)])


plt.xlabel("X")
plt.ylabel("Y")
plt.title("Représentation des données")
plt.show()







```

```python
# Catégorie du point (5,4)

print(svm.predict([[5,4]]))
```
